# Univariate Analysis
Univariate analysis is the simplest form of data analysis where data is being analyzed contains only one variable

```python
#plot Histogram
plt.figure(figsize=(10,8))
plt.hist(data['Ship Mode'])
plt.xticks(fontsize=10)
plt.show()
```

# In above graph the Standard Class is used most of the time.

```python
plt.figure(figsize=(16,12))
plt.hist(data['State'])
plt.xticks(rotation=80, fontsize=14)
plt.show()
```

# In above histogram graph we can see that  Kentucky, California,Florida are most frequently selling product states.

```python
plt.figure(figsize=(10,8))
plt.hist(data['Category'])
plt.xticks(fontsize=10)
plt.show()
```

# In above graph represents Office Supplies category of products are most selling as compard to others.


```python
plt.figure(figsize=(14,10))
plt.hist(data['Sub-Category'])
plt.xticks(rotation=80,fontsize=18)
plt.show()
```

# In above graph shows that Appliances, Paper,Accessories are most selling as compare to others.

```python
plt.figure(figsize=(10,8))
sns.countplot(x="Category",palette="rainbow",data=data)
plt.xticks(fontsize=12)
plt.xlabel("Category",fontsize=14)
plt.show()
```

# Data Visualization

```python
sns.pairplot(data)
```

```python
fig,axes=plt.subplots(1,1,figsize=(12,7))
sns.heatmap(data.corr())
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total profit vs Sales")
sns.barplot(data=data.groupby("Sub-Category")['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axes[1])
data.groupby("Sub-Category")['Sales','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=80,fontsize=12)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total Sales vs Quantity")
sns.barplot(data=data.groupby("Sub-Category")['Sales','Quantity'].agg(sum),x='Sales',y='Quantity',ax=axes[1])
data.groupby("Sub-Category")['Sales','Quantity'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=80,fontsize=12)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
data.groupby("Sub-Category")['Discount','Profit'].agg(sum).plot(kind='bar',ax=axes[1]).set_title('Discount and Profit Relation based on Sub_category')
data.groupby("Sub-Category")['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Quantity and Profit Relation based on Sub_category')
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Distributution plots",fontsize=16)
sns.distplot(data['Sales'],ax=axes[0,0])
sns.distplot(data['Profit'],ax=axes[0,1])
sns.distplot(data['Discount'],ax=axes[1,0])
sns.distplot(data['Quantity'],ax=axes[1,1])
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Sales with different shipping modes and segments",fontsize=16)
sns.barplot(data['Ship Mode'],data['Sales'],ax=axes[0,0])
sns.lineplot(data['Ship Mode'],data['Sales'],ax=axes[0,1])
sns.barplot(data['Segment'],data['Sales'],ax=axes[1,0])
sns.lineplot(data['Segment'],data['Sales'],ax=axes[1,1])
plt.show()
```

```python
fig,ax=plt.subplots(1,1,figsize=(12,8))
sns.countplot(data['Quantity'],hue=data['Region'])
plt.show()
```
